## Approach #1

```python
class Solution:
    def isPalindrome(self, s: str) -> bool:
        temp = []
        alpha_numeric = ['a','b','c','d','e','f','g','h','i','j','k','l','m','n','o','p','q','r','s','t','u','v','w','x','y','z','1','2','3','4','5','6','7','8','9','0']
        for i in s.lower():
            if i in alpha_numeric:
                temp.append(i)
        if temp == temp[::-1]:
            return True
        else:
            return False
```

## Approach #2

```python
class Solution:
    def isPalindrome(self, s: str) -> bool:
        temp = []
        for i in s.lower():
            if i.isalnum():
                temp.append(i)
        if temp==temp[::-1]:
            return True
        else:
            return False
        
```
